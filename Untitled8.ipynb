```python
import matplotlib.pyplot as plt
from PIL import Image

def crop_image(input_path, output_path, left, upper, right, lower):
    # Buka gambar
    image = Image.open(input_path)

    # Crop gambar
    cropped_image = image.crop((left, upper, right, lower))

    # Simpan gambar yang sudah di-crop
    cropped_image.save(output_path)

    # Tampilkan gambar hasil crop
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(image)
    axes[0].axis('off')
    axes[0].set_title('Gambar Asli')
    axes[1].imshow(cropped_image)
    axes[1].axis('off')
    axes[1].set_title('Hasil Crop')
    plt.show()

    print("Gambar berhasil di-crop dan disimpan di", output_path)

# Path file input dan output gambar pertama
input_path1 = "123.jpg"
output_path1 = "123.jpg"

# Koordinat crop untuk gambar pertama (misalnya)
left1 = 100
upper1 = 100
right1 = 400
lower1 = 400

# Panggil fungsi crop_image untuk gambar pertama
crop_image(input_path1, output_path1, left1, upper1, right1, lower1)

# Path file input dan output gambar kedua
input_path2 = "Messi.jpg"
output_path2 = "Messi.jpg"

# Koordinat crop
```

```python
import cv2
import matplotlib.pyplot as plt

# Mendefinisikan output path
output_path = "123.jpg"

# Membaca citra menggunakan OpenCV
image = cv2.imread(output_path, cv2.IMREAD_GRAYSCALE)

# Mengubah citra menjadi negatif
negative_image = cv2.bitwise_not(image)

# Menghitung histogram
hist = cv2.calcHist([negative_image], [0], None, [256], [0, 256])

# Menampilkan citra asli dan citra negatif
plt.subplot(2, 1, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Citra Asli")
plt.axis('off')

plt.subplot(2, 1, 2)
plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
plt.title("Citra Negatif")
plt.axis('off')

plt.show()

# Menampilkan histogram
plt.plot(hist, color='black')
plt.title("Histogram Citra Negatif")
plt.xlabel("Nilai Pixel")
plt.ylabel("Frekuensi")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Mendefinisikan output path
output_path = "123.jpg"

# Membaca citra menggunakan OpenCV
image = cv2.imread(output_path, cv2.IMREAD_GRAYSCALE)

# Mengubah citra menjadi negatif
negative_image = cv2.bitwise_not(image)

# Menambahkan brightness ke citra negatif
brightness_value = 50
brightened_image = cv2.add(negative_image, brightness_value)

# Menghitung histogram citra negatif
negative_hist = cv2.calcHist([negative_image], [0], None, [256], [0, 256])

# Menghitung histogram citra brightness
brightness_hist = cv2.calcHist([brightened_image], [0], None, [256], [0, 256])

# Menampilkan citra asli, citra negatif, dan citra brightness
plt.subplot(3, 1, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Citra Asli")
plt.axis('off')

plt.subplot(3, 1, 2)
plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
plt.title("Citra Negatif")
plt.axis('off')

plt.subplot(3, 1, 3)
plt.imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
plt.title("Citra Brightness")
plt.axis('off')

plt.show()

# Menampilkan histogram citra negatif
plt.plot(negative_hist, color='black')
plt.title("Histogram Citra Negatif")
plt.xlabel("Nilai Pixel")
plt.ylabel("Frekuensi")
plt.show()

# Menampilkan histogram citra brightness
plt.plot(brightness_hist, color='black')
plt.title("Histogram Citra Brightness")
plt.xlabel("Nilai Pixel")
plt.ylabel("Frekuensi")
plt.show()
```
